```python
import pandas as pd

# 1. EXTRACT: Load the CSV files into Pandas DataFrames
print("Loading data...")
orders = pd.read_csv('olist_orders_dataset.csv')
items = pd.read_csv('olist_order_items_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')

# 2. TRANSFORM: Merge the tables together (Just like SQL JOINs!)
print("Merging datasets...")
# First, join items to orders on order_id
merged_df = pd.merge(items, orders, on='order_id', how='inner')
# Next, join the products table on product_id
final_df = pd.merge(merged_df, products, on='product_id', how='inner')

# 3. TRANSFORM: Data Cleaning
# Drop any rows where the product category is missing (NaN)
cleaned_df = final_df.dropna(subset=['product_category_name'])

# 4. ADVANCED WRANGLING: GroupBy and Aggregation
# Let's recreate yesterday's SQL revenue calculation in Python!
category_revenue = cleaned_df.groupby('product_category_name')['price'].sum().reset_index()

# Sort the values in descending order to find the top categories
top_categories = category_revenue.sort_values(by='price', ascending=False).head(5)

# Format the output to look like currency using a lambda function
top_categories['price'] = top_categories['price'].apply(lambda x: f"${x:,.2f}")

# 5. LOAD/VIEW: Display the final result
print("\n--- Top 5 Product Categories by Revenue ---")
display(top_categories)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. LOAD: Read the items dataset
items = pd.read_csv('olist_order_items_dataset.csv')

# 2. DESCRIPTIVE STATISTICS: Get the math summary
print("--- Basic Statistics for Price and Freight Value ---")
display(items[['price', 'freight_value']].describe())

# 3. DISTRIBUTION PLOT (Histogram)
plt.figure(figsize=(10, 5))
# We limit the x-axis to 500 to zoom in on the majority of orders
sns.histplot(items['price'], bins=50, kde=True, color='teal')
plt.title('Distribution of Product Prices')
plt.xlabel('Price (Brazilian Real)')
plt.ylabel('Number of Orders')
plt.xlim(0, 500)
plt.show()

# 4. CORRELATION HEATMAP
plt.figure(figsize=(6, 4))
# Calculate the mathematical correlation between price and shipping
correlation_matrix = items[['price', 'freight_value']].corr()

# Draw the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation: Price vs. Shipping Cost')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt

# 1. LOAD & MERGE
orders = pd.read_csv('olist_orders_dataset.csv')
customers = pd.read_csv('olist_customers_dataset.csv')

# Link the order to the unique human who bought it
df = pd.merge(orders, customers, on='customer_id')

# 2. CLEANING: Keep only delivered orders and necessary columns
df = df[df['order_status'] == 'delivered']
df = df[['customer_unique_id', 'order_purchase_timestamp']]

# 3. FEATURE ENGINEERING: Date Formatting
# Convert text strings into actual Python datetime objects
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])

# Create 'Order_Month' (Force the date to the 1st of the month for grouping)
df['Order_Month'] = df['order_purchase_timestamp'].dt.to_period('M').dt.to_timestamp()

# Create 'Cohort_Month' (Find the absolute minimum purchase date per unique customer)
df['Cohort_Month'] = df.groupby('customer_unique_id')['Order_Month'].transform('min')

# 4. GROUPING: Count unique customers per cohort per month
cohort_data = df.groupby(['Cohort_Month', 'Order_Month'])['customer_unique_id'].nunique().reset_index()

# Calculate the 'Cohort Index' (Months elapsed since first purchase)
# Example: 0 = First month, 1 = Returned the next month, etc.
cohort_data['Cohort_Index'] = (
    (cohort_data['Order_Month'].dt.year - cohort_data['Cohort_Month'].dt.year) * 12
    + (cohort_data['Order_Month'].dt.month - cohort_data['Cohort_Month'].dt.month)
)

# 5. TRANSFORM: Pivot the table into a matrix format
cohort_counts = cohort_data.pivot(index='Cohort_Month', columns='Cohort_Index', values='customer_unique_id')

# Divide the columns by the initial cohort size (Month 0) to get percentage rates
cohort_sizes = cohort_counts.iloc[:, 0]
retention = cohort_counts.divide(cohort_sizes, axis=0)

# 6. VISUALIZE: Plot the Retention Heatmap
plt.figure(figsize=(14, 8))
# We format the numbers as percentages and cap the color scale at 1% for visibility
sns.heatmap(retention, annot=True, fmt='.1%', cmap='Blues', vmin=0.0, vmax=0.01)
plt.title('Customer Retention Matrix (Olist E-Commerce)')
plt.ylabel('Cohort Month')
plt.xlabel('Months Since First Purchase')
plt.show()
```
